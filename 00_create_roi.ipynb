```python
import geopandas as gpd
from shapely.geometry import box

# Define bounding box coordinates (Punjabâ€“Delhi region)
west, south, east, north = 75.0, 28.0, 77.8, 31.5

# Create polygon
roi_geom = box(west, south, east, north)

# Create GeoDataFrame
roi_gdf = gpd.GeoDataFrame(
    {"name": ["Punjab-Delhi_Agriculture"]},
    geometry=[roi_geom],
    crs="EPSG:4326"
)

# Save as GeoJSON
roi_gdf.to_file("roi.geojson", driver="GeoJSON")

print("roi.geojson created successfully")
```
